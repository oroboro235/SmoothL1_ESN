```python
outputs_path = "./results/outputs/"
models_path = "./results/models/"

# mackey-glass
mg_none_preds = np.load(outputs_path + "mg_none_preds.npy")
mg_none_trues = np.load(outputs_path + "mg_none_trues.npy")
mg_l2_preds = np.load(outputs_path + "mg_l2_preds.npy")
mg_l2_trues = np.load(outputs_path + "mg_l2_trues.npy")
mg_smoothl1_preds = np.load(outputs_path + "mg_smoothl1_preds.npy")
mg_smoothl1_trues = np.load(outputs_path + "mg_smoothl1_trues.npy")

# lorenz
lorenz_none_preds = np.load(outputs_path + "lorenz_none_preds.npy")
lorenz_none_trues = np.load(outputs_path + "lorenz_none_trues.npy")
lorenz_l2_preds = np.load(outputs_path + "lorenz_l2_preds.npy")
lorenz_l2_trues = np.load(outputs_path + "lorenz_l2_trues.npy")
lorenz_smoothl1_preds = np.load(outputs_path + "lorenz_smoothl1_preds.npy")
lorenz_smoothl1_trues = np.load(outputs_path + "lorenz_smoothl1_trues.npy")


```

```python
# mg plot
plt.plot(mg_smoothl1_preds)
plt.plot(mg_smoothl1_trues)
plt.show()
```

```python
# lorenz plot 3d
def plot_3d(preds, trues):
    X_preds = preds[:, 0]; y_preds = preds[:, 1]; z_preds = preds[:, 2]
    X_trues = trues[:, 0]; y_trues = trues[:, 1]; z_trues = trues[:, 2]
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X_preds, y_preds, z_preds, c='r', marker='o', label='Predictions')
    ax.scatter(X_trues, y_trues, z_trues, c='b', marker='^', label='Ground Truth')
    ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
    ax.legend()
    plt.show()
    
    return fig, ax

# none
plot_3d(lorenz_none_preds, lorenz_none_trues)
# l2
plot_3d(lorenz_l2_preds, lorenz_l2_trues)
# smoothl1
plot_3d(lorenz_smoothl1_preds, lorenz_smoothl1_trues)
```
